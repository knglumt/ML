```python
# imports
import pandas as pd
import numpy as np
```

```python
# Read data from CSV file 
df=pd.read_csv("car-pricing-data.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# drop nulls
df.dropna(inplace=True) 
df.isnull().sum()
```

```python
# convert mileage, engine, max_power to number
df[['mileage','engine','max_power']] = df[['mileage','engine','max_power']].replace(to_replace=r'([a-z/]+|[A-Z/]+)', value='', regex=True)
df[['mileage','engine','max_power']] = df[['mileage','engine','max_power']].replace(to_replace=' ', value='', regex=True)
df[['mileage','engine','max_power']] = df[['mileage','engine','max_power']].apply(pd.to_numeric)
```

```python
df
```

```python
#drop duplicates rows
df.drop_duplicates()
```

```python
# count value from coloumns
print(df['name'].value_counts())
print(df['seller_type'].value_counts())
print(df['transmission'].value_counts())
print(df['fuel'].value_counts())
print(df['seats'].value_counts())
print(df['owner'].value_counts())
```

```python
# Find the correlation between features (hints: pearson correlation)
df.corr(method="pearson")
```

```python
#Draw heatmap example
cm=df.corr(method="pearson")
import seaborn as sns
sns.set(font_scale=0.9)
hm=sns.heatmap(cm, cbar=True,annot=True,square=True)
```

```python
# Split the data to independent variables & dependent variable
dfreg=df[['year','km_driven','mileage','engine','max_power','selling_price']]
dfreg.head()
```

```python
# all features; year, km_driven, mileage, engine, max_power 
x1 = dfreg.iloc[:,0:5].values
# select year, km_driven, engine, max_power; F2
x2 = dfreg[['year', 'km_driven', 'engine', 'max_power']]
# select year, engine, max_power, mileage; F3
x3 = dfreg[['year', 'engine', 'max_power', 'mileage']]

y = dfreg.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
# Split the data to training and test sets
# F1
x_train1,x_test1, y_train1, y_test1=train_test_split(x1,y,test_size=0.2,random_state=147)
# F2
x_train2,x_test2, y_train2, y_test2=train_test_split(x2,y,test_size=0.2,random_state=147)
# F3
x_train3,x_test3, y_train3, y_test3=train_test_split(x3,y,test_size=0.2,random_state=147)
```

```python
# Print the number of instances in training & test set
x_train1.shape
```

```python
y_train1.shape
```

```python
x_test1.shape
```

```python
# Initialize min-max scaler and transform each feature by using min-max scaler
# You need to put the feature values to a certain range (in general: (0, 1)) in order to stabilize the model
from sklearn.preprocessing import MinMaxScaler
scaler1=MinMaxScaler(feature_range=(0,1))
scaler2=MinMaxScaler(feature_range=(0,1))
scaler3=MinMaxScaler(feature_range=(0,1))
```

```python
x_train1=scaler1.fit_transform(x_train1)
x_train2=scaler2.fit_transform(x_train2)
x_train3=scaler3.fit_transform(x_train3)
```

```python
x_train1
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Initialize the linear regression model
model1=LinearRegression()
model2=LinearRegression()
model3=LinearRegression()
```

```python
# Fit the training data to the model
model1.fit(x_train1, y_train1)
model2.fit(x_train2, y_train2)
model3.fit(x_train3, y_train3)
```

```python
model1.coef_
model2.coef_
model3.coef_
```

```python
model1.intercept_
model2.intercept_
model3.intercept_
```

```python
# Print the general formula of our linear regression model
_str="y= "
for i, m in enumerate(model1.coef_):
    _str += "x_{}*{}+".format(i+1,m)
_str += str(model1.intercept_)
print(_str)
```

```python
# Scale each feature to range(0, 1)
x_test1=scaler1.transform(x_test1)
x_test2=scaler2.transform(x_test2)
x_test3=scaler3.transform(x_test3)
x_test1[0:5]
```

```python
x_train1[0:5]
```

```python
x_test1[0:5]
```

```python
# set NaN values as zero
x_test1=np.nan_to_num(x_test1)
x_test2=np.nan_to_num(x_test2)
x_test3=np.nan_to_num(x_test3)
```

```python
np.where(np.isnan(x_test1))
np.where(np.isnan(x_test2))
np.where(np.isnan(x_test3))
```

```python
# Predict the values by using all test data
y_predict1=model1.predict(x_test1)
y_predict2=model2.predict(x_test2)
y_predict3=model3.predict(x_test3)
```

```python
# Print the predicted and the actual value of the data in test set
#F1
for i in range(len(y_test1)):
    print("Predicted: ",y_predict1[i]," Actual:", y_test1[i])
#F2
for i in range(len(y_test2)):
    print("Predicted: ",y_predict2[i]," Actual:", y_test2[i])
#F3 
for i in range(len(y_test3)):
    print("Predicted: ",y_predict3[i]," Actual:", y_test3[i])
```
